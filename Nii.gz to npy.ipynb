```python
import os
import numpy as np
import nibabel as nib

# 입력 폴더와 출력 폴더 경로
input_folder = '/SSD3_8TB/Daniel/dataset/IXI/IXI-T1'
output_folder = '/SSD3_8TB/Daniel/dataset/IXI/IXI-T1-npy'

# 출력 폴더가 없으면 생성
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 입력 폴더에서 .nii.gz 파일 목록 가져오기
nii_files = [f for f in os.listdir(input_folder) if f.endswith('.nii.gz')]

# 각 파일에 대해 .npy 파일로 변환하고 저장
for nii_file in nii_files:
    # .nii.gz 파일 읽기
    nii_path = os.path.join(input_folder, nii_file)
    nii_data = nib.load(nii_path).get_fdata()
    
    # .npy 파일로 저장
    npy_file = nii_file.replace('.nii.gz', '.npy')
    npy_path = os.path.join(output_folder, npy_file)
    np.save(npy_path, nii_data)

print("모든 .nii.gz 파일이 .npy 파일로 변환되었습니다.")
```

```python
import os
import nibabel as nib

# 입력 폴더 경로
input_folder = '/SSD3_8TB/Daniel/dataset/IXI/IXI-T1'

# 입력 폴더에서 .nii.gz 파일 목록 가져오기
nii_files = [f for f in os.listdir(input_folder) if f.endswith('.nii.gz')]

# 각 파일에 대해 numpy 배열로 읽고 shape을 저장할 집합 자료구조 사용
shape_set = set()

for nii_file in nii_files:
    # .nii.gz 파일 읽기
    nii_path = os.path.join(input_folder, nii_file)
    nii_data = nib.load(nii_path).get_fdata()
    
    # shape 저장
    shape_set.add(nii_data.shape)

# 서로 다른 shape 종류와 개수 출력
print(f"서로 다른 shape의 개수: {len(shape_set)}")
print("Shape 종류:")
for shape in shape_set:
    print(shape)
```
